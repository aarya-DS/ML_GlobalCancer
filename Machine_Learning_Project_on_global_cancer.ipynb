1. Reading the dataset from kaggle

```python
import pandas as pd
df=pd.read_csv("/content/GlobalCancer.csv")
df
```

Generating a basic plot

```python
import matplotlib.pyplot as plt

# Calculate the frequency of each cancer stage
cancer_stage_counts = df['Cancer_Stage'].value_counts()

# Create the bar plot using the cancer stages and their frequencies
plt.bar(cancer_stage_counts.index, cancer_stage_counts.values,color='blue')
plt.xlabel("Cancer Stage")
plt.ylabel("Frequency")
plt.title("Distribution of Cancer Stages")
plt.show()
```

2. Cleaning the data i.e, removing unnecessary columns

```python
df = df.drop(['Patient_ID', 'Year'], axis = 1)
df
```

Checking if there are any NULL values

```python
df.isnull().sum()
```

3. Encoding the data , converting categorical data(string) ---> numerical data

Each column with string are given a particular value of numbers, each values states a number

```python
# prompt: convert catergorical data into numerical data using label encoding , for all columns in dataset

from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame as defined in the previous code

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Iterate through each column in the DataFrame
for column in df.columns:
    # Check if the column's data type is categorical (object)
    if df[column].dtype == 'object':
        # Fit and transform the column using LabelEncoder
        df[column] = label_encoder.fit_transform(df[column])

# Print the updated DataFrame with numerical data
df
```

In Machine Learning we use the datasets with numbers only

```python
X = df.drop("Treatment_Cost_USD", axis =1)
y = df['Treatment_Cost_USD']
```

```python
X # contains all values except Cancer_stage
```

```python
y #contains only cancer_stage column
```

5. Train-Test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#X_train
X_test
```

```python
X_train #contains all columns except Treatment cost (80%)
```

```python
y_train #contains treatment cost(20%)
```

```python
X_test #contains all columns except treatment cost (20%)
```

```python
y_test #contains treatment cost column(20%)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 6. Model Selection and Training
# We will use Linear Regression for this task as it's a suitable model for predicting continuous values (Treatment_Cost_USD)
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)

# 7. Prediction and Evaluation
# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Treatment Cost (USD)")
plt.ylabel("Predicted Treatment Cost (USD)")
plt.title("Actual vs. Predicted Treatment Cost")
plt.grid(True)
plt.show()
```

```python

import plotly.graph_objects as go
import numpy as np

errors = y_test - y_pred

fig = go.Figure(data=go.Scatter(x=y_test, y=errors, mode='markers'))

# Add a horizontal line at y=0 to represent zero error
fig.add_shape(type="line",
    x0=min(y_test), y0=0, x1=max(y_test), y1=0,
    line=dict(color="Red", width=2)
)

fig.update_layout(
    title='Errors (Actual - Predicted) vs. Actual Treatment Cost',
    xaxis_title='Actual Treatment Cost (USD)',
    yaxis_title='Error (USD)'
)

fig.show()
```
